OUTLIERS DETECTION AND TREATMENT

```python
#detect outliers using boxplot
sns.boxplot(df['Ride Distance'])
plt.title('Boxplot of Ride Distance')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = ["Booking Value", "Ride Distance", "Driver Ratings", "Customer Rating"]

for col in num_cols:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
#Detect outliers using IQR(Inter Quartile Range) method
Q1 = df["Booking Value"].quantile(0.25)
Q3 = df["Booking Value"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outliers = df[(df["Booking Value"] < lower_bound) | (df["Booking Value"] > upper_bound)]
print("Number of outliers in Booking Value:", outliers.shape[0])
```

```python
df_removed = df[(df["Booking Value"] >= lower_bound) & (df["Booking Value"] <= upper_bound)]
print("Shape after removing outliers:", df_removed.shape)
```

```python
df_capped = df.copy()
df_capped["Booking Value"] = df_capped["Booking Value"].clip(lower=0, upper=1371.5)

print("Max Booking Value after capping:", df_capped["Booking Value"].max())
print("Shape after capping:", df_capped.shape)
```

What is the average booking value for each payment method?

```python
# compute counts and mean
grp = (df_viz[["Payment Method","Booking Value"]]
       .dropna()
       .groupby("Payment Method")
       .agg(count=("Booking Value","size"), mean=("Booking Value","mean"))
       .sort_values("mean", ascending=False))

print(grp)

# barplot ordered by mean
order = grp.index.tolist()
plt.figure(figsize=(10,5))
sns.barplot(x=grp.index, y=grp["mean"].values, palette="Blues_d")
plt.xticks(rotation=45)
plt.title("Average Booking Value by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Average Booking Value (â‚¹)")
# annotate bars with mean values
for i, val in enumerate(grp["mean"].values):
    plt.text(i, val + val*0.02, f"{val:,.0f}", ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df_viz, x="Payment Method", y="Booking Value", order=order)
plt.xticks(rotation=45)
plt.title("Booking Value distribution by Payment Method")
plt.ylabel("Booking Value (â‚¹)")
plt.xlabel("Payment Method")
plt.ylim(0, df_viz["Booking Value"].quantile(0.99))   # optional zoom to 99th pct
plt.tight_layout()
plt.show()
```

What is the average Booking Value for each Vehicle Type?

```python
grp = (df_viz[['Vehicle Type', 'Booking Value']]
       .dropna()
       .groupby('Vehicle Type')
       .agg(count=('Booking Value', 'size'), mean=('Booking Value', 'mean'))
       .sort_values('mean', ascending=False))

print(grp)
```

```python
#Plotting a barplot
plt.figure(figsize=(10,5))
sns.barplot(x=grp.index, y = grp['mean'].values, palette = 'Greens_d')
plt.xticks(rotation=45)
plt.title('Average Booking Value by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.tight_layout()
plt.show()
```

```python
# Choose vehicle types with enough samples (top N)
top_vehicles = df_viz["Vehicle Type"].value_counts().head(8).index.tolist()
data = df_viz[df_viz["Vehicle Type"].isin(top_vehicles) & df_viz["Ride Distance"].notna()]
plt.figure(figsize=(14,6))
sns.boxplot(x="Vehicle Type", y="Ride Distance", data=data, order=top_vehicles)
plt.xticks(rotation=45)
plt.ylim(0, data["Ride Distance"].quantile(0.95))
plt.title("Ride Distance distribution by Vehicle Type")
plt.ylabel("Ride Distance (km)")
plt.xlabel("Vehicle Type")
plt.tight_layout()
plt.show()

```

```python
# Compute average ride distance per vehicle type
grp = (data.groupby("Vehicle Type")["Ride Distance"]
       .mean()
       .sort_values(ascending=False))

# Plot barplot
plt.figure(figsize=(14,6))
sns.barplot(x=grp.index, y=grp.values, palette="Blues_d")
plt.xticks(rotation=45)
plt.title("Average Ride Distance by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Average Ride Distance (km)")

# Annotate each bar with the value
for i, val in enumerate(grp.values):
    plt.text(i, val + val*0.02, f"{val:.1f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

Are Driver Ratings higher for certain Vehicle Types?

```python
# filter to top vehicle types
data_ratings = df_viz[df_viz["Vehicle Type"].isin(top_vehicles) & df_viz["Driver Ratings"].notna()]

plt.figure(figsize=(14,6))
sns.boxplot(x="Vehicle Type", y="Driver Ratings", data=data_ratings, order=top_vehicles, palette="Pastel1")
plt.xticks(rotation=45)
plt.title("Driver Ratings by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Driver Rating")
plt.ylim(0,5)   # ratings usually 1-5
# Also show mean points
means = data_ratings.groupby("Vehicle Type")["Driver Ratings"].mean().reindex(top_vehicles)
for i, m in enumerate(means):
    plt.text(i, m + 0.03, f"{m:.2f}", ha='center', fontsize=9, color='black')
plt.show()
```

Is Customer Rating better when Booking Status = Completed vs Cancelled/Incomplete?

```python
# boxplot across Booking Status
data_status = df_viz[df_viz["Booking Status"].notna() & df_viz["Customer Rating"].notna()]

plt.figure(figsize=(10,6))
sns.boxplot(x="Booking Status", y="Customer Rating", data=data_status, order=sorted(data_status["Booking Status"].unique()), palette="Set2")
plt.title("Customer Rating by Booking Status")
plt.xlabel("Booking Status")
plt.ylabel("Customer Rating")
plt.ylim(0,5)
plt.show()

# Compute means and counts
summary = (data_status.groupby("Booking Status")["Customer Rating"]
           .agg(count="size", mean="mean", median="median")
           .sort_values("mean", ascending=False))
print(summary)

# Optional: compare Completed vs Non-completed
data_status["is_completed"] = (data_status["Booking Status"] == "Completed")
means = data_status.groupby("is_completed")["Customer Rating"].agg(count="size", mean="mean")
print("\nCompleted vs Not Completed summary:")
print(means)
```
